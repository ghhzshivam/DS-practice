```python
# Recurring char
# this solution have O(n) time complexity and O(n) space complextity

a = [1]
l = []
for i in range(0,len(a)):
    if a[i] in l:
        print(a[i])
        break
    elif i == len(a)-1:
        print('Undefined')
    else:
        l.append(a[i])
        
    
    
```
