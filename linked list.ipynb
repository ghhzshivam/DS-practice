### __init__ is the constructor which contruct the function 
Self is the object calling name itself

```python
# Creatin the nodes
class node:
#     here init is the constructer and self is the object
    def __init__(self, data):
        self.data = data
        self.next = None

        
class LinkedList:
#     for creating the head
    def __init__(self):
        self.head = None
        
    def printlist(self):
        temp = self.head
        while(temp):
            print(temp.data)
            temp  =temp.next

            
if __name__=='__main__':
    linked_list = LinkedList()
    
#     here we ae creating the nodes for linkedlist
    linked_list.head = node(1)
    second = node(2)
    third = node(3)

# we are assigning the node and connecten thenm
linked_list.head.next = second
second.next = third 
```

```python
# Traversing the linked list
linked_list.printlist()
```
